```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 1: Load Data

```python
df = pd.read_csv('cleaned_data.csv')
```

Step 2: Check ride count and percentage of rides by both members and casual riders

```python
# Number of rides per user type
ride_counts = df['member_casual'].value_counts()
print("Total Rides by User Type:")
print(ride_counts)
```

```python
# Percentage of total rides per user type
ride_percentage = (df['member_casual'].value_counts(normalize=True) * 100).round(2)
print("\nPercentage of Total Rides:")
print(ride_percentage)
```

Step 3: Check trip duration in minutes for both members and casual riders

```python
# Average/Median trip duration grouped by user type
# Average and Median ride length grouped by user type
ride_length_stats = df.groupby('member_casual')['ride_length'].agg(['mean', 'median']).round(2)
print("Ride Length Statistics (in Minutes):")
print(ride_length_stats)
```

Step 4: Check trips by day of week grouped by user types

```python
# Set the correct order of days since python will sort them by alphabet order
day_order = ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
```

```python
# Total rides per day grouped by user type
rides_per_day = df.groupby(['member_casual', 'day_of_week']).size().unstack(level=0).reindex(day_order)
print("Total Rides per Day:")
print(rides_per_day)
```

```python
# Average rides per day grouped by user type
duration_per_day = df.groupby(['member_casual', 'day_of_week'])['ride_length'].mean().unstack(level=0).reindex(day_order).round(2)
print("\nAverage Ride Length (Minutes) per Day:")
print(duration_per_day)
```

Step 5: Check Hourly and Monthly Trips

```python
# Busiest hours grouped by user type
hourly_rides = df.groupby(['hour', 'member_casual']).size().unstack()

# Busiest months grouped by user type
monthly_rides = df.groupby(['month', 'member_casual']).size().unstack()
```

```python
print("\nHourly Rides per Day:")
print(hourly_rides)
```

```python
print("\nRides by Month:")
print(monthly_rides)
```

Step 6: Check for Popular Starting Stations

```python
# Filter out 'Unknown' stations
known_stations_df = df[df['start_station_name'] != 'Unknown']
```

```python
# Top Starting Stations for Casual Riders
print("Top 5 Start Stations for Casual Riders:")
print(known_stations_df[known_stations_df['member_casual'] == 'casual']['start_station_name'].value_counts().head(5))
```

```python
# Top Starting Stations for Member Riders
print("\nTop 5 Start Stations for Member Riders:")
print(known_stations_df[known_stations_df['member_casual'] == 'member']['start_station_name'].value_counts().head(5))
```
