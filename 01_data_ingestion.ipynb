```python
import pandas as pd
import glob
import os
```

Step 1: Locate the CSV Files

```python
# Set path to the raw data folder
path = './raw_data'

# Retrieve CSV files that match glob pattern
all_files = glob.glob(os.path.join(path, "*.csv"))
```

```python
# Verify file count is 12 (one for each month)
print(f"File Count: {len(all_files)}")
for file in all_files:
    print(os.path.basename(file)) #Output 12 for the file count and a list of the 12 files
```

Step 2: Loop to Combine CSV Files 

```python
# Empty list to store the dataframes
df_list = []

# Loop through and append each CSV file to the list
for file in all_files:
    df = pd.read_csv(file)
    df_list.append(df)
```

```python
# Concatenate monthly dataframes into one large yearly dataframe
combined_df = pd.concat(df_list, ignore_index=True)
```

Step 3: Check Data

```python
# Check shape to verify the concatenation worked
print(f"Data Shape: {combined_df.shape}")
```

```python
# Check the first few records to verify the columns are correctly aligned
combined_df.head()
```

Step 4: Export the Newly Created df to a CSV file

```python
# Path to save the CSV file
output_path = 'combined_raw_data.csv'
combined_df.to_csv(output_path, index=False)
```
